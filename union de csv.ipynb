```python
import pandas as pd
```

```python
df1 = pd.read_csv('uefa_club_coefficients_2013_2018.csv')
df2 = pd.read_csv('uefa_club_coefficients_2018_2023.csv')
```

```python
df1.head()
```

```python
df2.head()
```

Para hacer una predicción de lo que pueda pasar este año, hay muchos equipos que simplemente ya no nos interesan, el resto los vamos a dropear

```python
equipos_en_pie = ['Real Madrid','Atlético Madrid','Arsenal',
                  'PSV Eindhoven','Bayern München','FC København',
                  'Napoli','Internazionale','Real Sociedad',
                  'Lazio','Borussia Dortmund','Paris Saint-Germain',
                  'Manchester City','RB Leipzig','FC Barcelona','FC Porto']
```

```python
filtered_df1 = df1[df1['Team'].isin(equipos_en_pie)]
    
filtered_df1
```

```python
filtered_df2 = df2[df2['Team'].isin(equipos_en_pie)]
    
filtered_df2
```

```python
filtered_df1.sort_values(by='Team', inplace=True)
filtered_df2.sort_values(by='Team', inplace=True)
```

```python
filtered_df1.to_csv('uefa_club_coefficients_2013_2018_filtered.csv', index=False)
filtered_df2.to_csv('uefa_club_coefficients_2018_2023_filtered.csv', index=False)
```
