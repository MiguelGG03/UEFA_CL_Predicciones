# Intento de scrapeo en una página de datos de la UEFA

```python
from bs4 import BeautifulSoup
import requests
```

```python
link = "https://www.flashscore.com/football/europe/champions-league/results/"
req = requests.get(link)
```

Aquí tenemos el HTML en crudo

```python
req.text
```

Generamos nuestro propio objeto dinámico que recree la interfaz **DOM** pero en Python

```python
soup = BeautifulSoup(req.text)
soup
```

A simple vista pueden parecer lo mismo, pero se interactúa mucho mejor con esta interfaz

```python
type(soup.select("title"))
```

Es un conjunto de resultados que contiene los tags que concuerdan con el nombre `title`, por tanto es una lista.

Veamos qué tipo tiene ese primer valor del conjunto:

```python
type(soup.select("title")[0])
```

Ahora, con `getText()`, podemos recuperar el contenido de esta __Tag__

```python
soup.select("title")[0].getText()
```

Despues de trastear con los divs, siguiendo esta lógica podemos programar un script que recupere todos los atributos de los metadatos:

```python
for meta in soup.select("meta"):
    for atributo, valor in meta.attrs.items():
        print(f"{atributo}: {valor}")
```
