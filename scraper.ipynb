```python
table = soup.find('table', class_='t1')
```

```python
teams_data = []
```

```python
for row in table.find_all('tr', class_='clubline'):
    # Extrae las celdas de la fila
    cols = row.find_all('td')
    
    # Extrae los datos necesarios de cada celda
    rank = cols[0].text.strip()
    team_name = cols[2].text.strip()
    country = cols[3].text.strip()
    season_scores = [col.text.strip() for col in cols[4:9]]
    total_points = cols[9].text.strip()
    
    # Agrega los datos a la lista
    teams_data.append({
        'Rank': rank,
        'Team': team_name,
        'Country': country,
        "18/19" : season_scores[0],
        "19/20" : season_scores[1],
        "20/21" : season_scores[2],
        "21/22" : season_scores[3],
        "22/23" : season_scores[4],
        'Total Points': total_points
    })
```
